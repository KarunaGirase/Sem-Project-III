```python
import pandas as pd

# Load the dataset
df = pd.read_csv('crop_data_uncleaned.csv')

# Display the initial number of rows
print(f"Initial number of rows: {len(df)}")
```

```python
# Identify numerical and categorical columns
numerical_cols = ['Temperature', 'Humidity', 'Soil_Type', 'Rainfall']
categorical_col = 'Crop'

# Fill missing values in numerical columns with the median
for col in numerical_cols:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Fill missing values in the categorical column with the mode
mode_val = df[categorical_col].mode()[0]
df[categorical_col].fillna(mode_val, inplace=True)
```

```python
# Check for remaining missing values
print("\nDataFrame info after cleaning:")
df.info()

# Display the final number of rows
print(f"\nFinal number of rows: {len(df)}")

# Optionally, save the cleaned dataset to a new CSV file
df.to_csv('crop_data_cleaned.csv', index=False)
```
